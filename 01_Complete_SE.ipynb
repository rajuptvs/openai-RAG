```python
# get all the menu items and the links in  the nav menu into dataframes

menu_items = soup.find_all("li", {"class": "o-nav--site__item"})

dataframes = {}

for menu_item in menu_items:
    top_link = menu_item.find("a", {"class": "o-nav--site__top-link"})
    if top_link:
        name = top_link.text.strip()
        df = pd.DataFrame(columns=["name", "link"])
        sub_links = menu_item.find_all("a", {"class": "gtm-nav--site"})
        for sub_link in sub_links:
            df = df.append({"name": sub_link.text.strip(), "link": sub_link["href"]}, ignore_index=True)
        dataframes[name] = df

# example usage: print links under "About the Program" menu item
print(dataframes["About the Program"])
```

```python
import textwrap

def prettyprint(text, width=110):
    # Split the input text into lines based on newline characters
    lines = text.split('\n')

    # Wrap each line individually
    wrapped_lines = [textwrap.fill(line, width=width) for line in lines]

    # Join the wrapped lines back together using newline characters
    wrapped_text = '\n'.join(wrapped_lines)

    return wrapped_text
```

## Let's get started with LLM's and LangChain

```python
from langchain.chains.question_answering import load_qa_chain
from langchain import HuggingFaceHub
from langchain.llms import HuggingFacePipeline
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline, AutoModelForSeq2SeqLM

model_id = 'google/flan-t5-large'
tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoModelForSeq2SeqLM.from_pretrained(model_id, load_in_8bit=True, device_map='auto')

pipeline = pipeline(
    "text2text-generation",
    model=model,
    tokenizer=tokenizer,
    max_length=1024,

)

local_llm = HuggingFacePipeline(pipeline=pipeline)
```

```python
chain = load_qa_chain(local_llm, chain_type="stuff")
query = "What are the specializations available for software engineering"
docs = db.similarity_search(query)
chain.run(input_documents=docs, question=query)
```
